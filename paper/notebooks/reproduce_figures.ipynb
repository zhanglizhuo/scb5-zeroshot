# Reproduce Core Figures and Tables
This notebook loads precomputed baseline results and draws a quick verification chart for Table 3 core values.

```python
from pathlib import Path
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
repo_root = Path.cwd().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd()
results_path = repo_root / 'results' / 'baseline_results.json'
with open(results_path, 'r', encoding='utf-8') as f:
    results = json.load(f)
df = pd.DataFrame(results['table3_core'])
df
```

```python
ax = df.plot(x='subset', y='hit_at_1', kind='bar', legend=False, figsize=(7,4), color='#2A78C4')
ax.set_ylabel('Hit@1 (%)')
ax.set_xlabel('Subset')
ax.set_title('Table 3 Core Hit@1')
plt.tight_layout()
plt.show()
```
